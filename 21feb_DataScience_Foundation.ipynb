# DATASET CREATION

```python
np.random.seed(42)
n_students = 200

df = pd.DataFrame({
    "Student_ID": np.arange(1, n_students + 1),
    "Study_Hours_per_Week": np.random.normal(15, 5, n_students).round(1),  # average , spread , no of data -> round it to 1 digit
    "Attendance_Percentage": np.random.randint(50, 101, n_students),  # lowest , highest , no of data to be generated(exact)
    "Assignment_Score": np.random.randint(40, 101, n_students), # same
    "Midterm_Score": np.random.randint(35, 101, n_students) # same
})
df.head()
```

```python
# d = {
#     "Student_ID" : np.arange(1,201),
#     "score" : np.random.randint(50,101,200)
# }
# print(pd.DataFrame(d))
```

```python
df["Final_Score"] = (
    0.3 * df["Assignment_Score"]+
    0.3 * df["Midterm_Score"]+
    0.4 * (df["Study_Hours_per_Week"] * 2)
).round(1)
```

```python
df["Grade"] = np.where(df["Final_Score"] >= 85, "A",
                np.where(df["Final_Score"] >= 70, "B",
                np.where(df["Final_Score"] >= 55, "C", "Fail")))


df.head()
```
