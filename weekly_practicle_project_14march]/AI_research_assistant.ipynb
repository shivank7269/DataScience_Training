```python
from google import genai
import os
from dotenv import load_dotenv

load_dotenv()
API_KEY = os.getenv('API_KEY')
client = genai.Client(api_key=API_KEY)

article_text = """
Renewable energy is energy that is collected from renewable resources that are naturally replenished on a human timescale, such as sunlight, wind, rain, tides, waves, and geothermal heat. Renewable energy often provides energy for electricity generation, solar water heating, and cooling. Although most renewable energy projects are large-scale, renewable technologies are also suited to rural and remote areas and developing countries, where energy is often crucial in human development.
The rapid deployment of renewable energy and energy efficiency, and technological diversification of energy sources, would result in significant energy security and economic benefits. It would also reduce environmental pollution such as air pollution caused by burning of fossil fuels and improve public health, reduce premature mortalities from pollution and save associated health costs that amount to several hundred billion dollars annually only in the United States.
"""

response = client.models.generate_content(
    model='gemini-2.5-flash',
    contents=f"Please summarize the following article:\n{article_text}"
)

print("Generated Output")
print(response.text)
```

Task 2 – Prompt Engineering Experiments (8 Marks)
Design and test three different prompting strategies.

1️⃣ Zero-Shot Prompt
Ask the model to summarize the article without providing examples.

Example task:

Summarize the following article in 5 bullet points.

2️⃣ Few-Shot Prompt
Provide 2 example summaries before asking the model to summarize a new article.

3️⃣ Chain-of-Thought Prompt
Ask the model to reason step-by-step before generating the summary.

Example structure:

Analyze the article step by step:

1. Identify the main topic

2. Extract key ideas

3. Generate a concise summary

Deliverable:

All three prompts


Model outputs


Comparison of responses

```python
prompt_zero_shot = """
Summarize the following article in exactly 5 bullet points. Focus on the main ideas and their impacts.

{article}
"""

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt_zero_shot.format(article=article_text)
)
print("Zero‑Shot Prompt Output:")
print(response.text)
```

```python
prompt_few_shot = """
Here are two example article summaries:

Example 1:
Article: "Machine learning is a subset of artificial intelligence that enables computers to learn from data without explicit programming. It is widely used in recommendation systems, fraud detection, and medical diagnosis."
Summary:
• Machine learning lets computers learn patterns from data without being explicitly programmed.
• It is used in applications such as recommendations, fraud detection, and healthcare diagnostics.
• The main advantage is the ability to adapt to new data and improve over time.

Example 2:
Article: "Blockchain is a distributed ledger technology that records transactions across many computers in a secure way. It ensures transparency, immutability, and trust in decentralized systems."
Summary:
• Blockchain stores data in a tamper‑resistant distributed ledger.
• It ensures transparency and immutability of records across multiple nodes.
• It enables trustless interactions in decentralized platforms.

Now, summarize the following new article in the same style (3–5 bullet points):

{article}
"""

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt_few_shot.format(article=article_text)
)
print("\nFew‑Shot Prompt Output:")
print(response.text)
```

```python
prompt_cot = """
Analyze the article step by step, then generate a concise summary.

Step 1: Identify the main topic of the article.
Step 2: Extract 3–5 key ideas or impacts.
Step 3: Write a 4–5 bullet‑point summary that captures the essence.

Now perform the analysis on the following article:

{article}
"""

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt_cot.format(article=article_text)
)
print("\nChain‑of‑Thought Prompt Output:")
print(response.text)

```

Task 3 – Prompt Optimization (7 Marks)
Improve the prompt so that the AI generates better executive summaries.

Your optimized prompt must produce:

3 key insights


1 actionable takeaway


Professional tone


Test the optimized prompt on two different articles.

Deliverable:

Final optimized prompt


Two example outputs

```python
from dotenv import load_dotenv
from google import genai
import os

load_dotenv()
API_KEY = os.getenv("API_KEY")
client = genai.Client(api_key=API_KEY)

def get_executive_summary(article):
    prompt = """
You are an executive research assistant. Read the article carefully and provide a short, professional executive summary.

Format your response exactly as follows:

**3 KEY INSIGHTS**
• [Insight 1]
• [Insight 2]
• [Insight 3]

**1 ACTIONABLE TAKEAWAY**
• [Clear, practical recommendation]

Requirements:
- Use a professional, concise tone.
- Base everything strictly on the article content.
- Do not add external facts or references.

Now analyze this article:

{article}
""".format(article=article)

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )
    return response.text


get_executive_summary(article_text)
```

Task 4 – Tokenization Experiment (5 Marks)
Use a tokenizer from HuggingFace or any LLM library.
Steps:
Tokenize a paragraph of text.


Count the total number of tokens.


Display the tokenized output.


Example tools:
tiktoken


transformers tokenizer


Deliverable:

Example tokenized output

```python
from transformers import AutoTokenizer

# Load a standard tokenizer (e.g., from a BERT‑like model)
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Paragraph to tokenize (you can change this)
text = """
Renewable energy is energy that is collected from renewable resources that are naturally replenished on a human timescale,
such as sunlight, wind, rain, tides, waves, and geothermal heat. Renewable energy often provides energy for electricity
generation, solar water heating, and cooling.
"""

# Step 1: Tokenize the text
tokens = tokenizer.tokenize(text)

# Step 2: Convert tokens to token IDs (optional but common)
token_ids = tokenizer.encode(text, add_special_tokens=True)

# Step 3: Count total tokens
token_count = len(token_ids)

# Step 4: Display tokenized output
print("Original text:")
print(text)
print("\nTokenized tokens:")
print(tokens)
print("\nToken IDs:")
print(token_ids)
print("\nTotal number of tokens:", token_count)
```

Task 5 – Build a Mini AI Tool (5 Marks)
Create a simple AI assistant that takes a long article and produces:

A short summary


Key insights


One actionable recommendation


The program should:

Accept text input


Send prompt to LLM


Display structured output


Deliverable:

Python script or notebook


Example result

```python
from dotenv import load_dotenv
from google import genai
import os

# Load .env file
load_dotenv()
API_KEY = os.getenv("API_KEY")
if not API_KEY:
    raise ValueError("API_KEY not found in .env file")

client = genai.Client(api_key=API_KEY)

def create_structured_prompt(article_text):
    return f"""
You are an executive research assistant. Analyze the following article and produce a structured response:

**1. Short Summary**
Write 2–3 concise sentences summarizing the article.

**2. Key Insights**
List 3–5 key insights (as bullet points) that capture the most important ideas.

**3. Actionable Recommendation**
Provide 1 clear, practical recommendation based on the article (single bullet point).

Requirements:
- Use professional tone and avoid adding external information.
- Keep insights specific to the article content.

Article:
{article_text}
"""

def generate_research_summary(article_text):
    prompt = create_structured_prompt(article_text)

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )
    return response.text

# Example usage (you can replace this with user input)
sample_article = """
Renewable energy is energy that is collected from renewable resources that are naturally replenished on a human timescale,
such as sunlight, wind, rain, tides, waves, and geothermal heat. Renewable energy often provides energy for electricity
generation, solar water heating, and cooling. Although most renewable energy projects are large‑scale, renewable technologies
are also suited to rural and remote areas and developing countries, where energy is often crucial in human development.

The rapid deployment of renewable energy and energy efficiency, and technological diversification of energy sources,
would result in significant energy security and economic benefits. It would also reduce environmental pollution such as
air pollution caused by burning of fossil fuels and improve public health, reduce premature mortalities from pollution
and save associated health costs that amount to several hundred billion dollars annually only in the United States.
"""

print("=== Mini AI Research Assistant ===\n")
print("Input article preview:")
print(sample_article[:200] + "...\n")

output = generate_research_summary(sample_article)
print("Structured AI Output:")
print(output)
```
