```python
import pandas as  pd
import numpy as np
```

```python
customer = pd.read_csv("../Dataset/ecommerce/Customer.csv")
sales = pd.read_csv("../Dataset/ecommerce/Sales.csv")
support = pd.read_csv("../Dataset/ecommerce/Support.csv")
```

```python
rows , columns = customer.shape
print(f"Customer Rows:{rows} Columns:{columns}")
rows , columns = sales.shape
print(f"Sales Rows:{rows} Columns:{columns}")
rows , columns = support.shape
print(f"Supportn Rows:{rows} Columns:{columns}")

print(f"Column Name for Customer :{list(customer.keys())}")
print(f"Column Name for Sales :{list(sales.keys())}")
print(f"Column Name for Support :{list(support.keys())}")

print(f"Missing Values in Customer :\n{customer.isnull().sum().sort_values(ascending=False).to_string()}")
print(f"Missing Values in Sales :\n{sales.isnull().sum().sort_values(ascending=False).to_string()}")
print(f"Missing Values in Support :\n{support.isnull().sum().sort_values(ascending=False).to_string()}")
```

```python
sales["Price"] = sales["Price"]*0.90
sales["Revenue"] = sales["Price"] * sales["Quantity"]
sales
```

```python
sales["OrderDate"] = pd.to_datetime(sales["OrderDate"])
sales[(sales["OrderDate"].dt.month==1) & (sales["OrderDate"].dt.year == 2023)]
sales.head(10)
```

```python
customer[customer["Region"] == "North"]
sales[sales["Revenue"] > 10000]
```

```python
customer.sort_values("SignupDate")
sales.sort_values("Revenue",ascending=False)
```

```python
customer["Age"] =customer["Age"].fillna(customer["Age"].median())
merged = pd.merge(customer,sales,on="CustomerID")
merged.groupby("Region")["Revenue"].mean()
support.groupby("IssueType")["ResolutionTime"].mean()
```

```python
# Data wrangling
customer["Age"] =customer["Age"].fillna(customer["Age"].median())
# customer.rename(columns={"name":"Name"}, inplace=True)
final_merge = pd.merge(merged,support,on = "CustomerID")
final_merge["CLV"] = final_merge.groupby("CustomerID")["Revenue"].transform("sum")
```

```python
final_merge.groupby("CustomerID")["ResolutionTime"].mean()
```

```python
final_merge.to_csv("cleaned_ecommerce_data.csv",index=False)
```
