```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score
from sklearn.model_selection import train_test_split

df = pd.read_csv('../Dataset/data/salary_lpa - Sheet1.csv')

x=df[['Experience_years']]
y=df['Salary_lpa']


x_train,x_test,y_train,y_test = train_test_split(
    x,y,test_size=0.2,random_state=42
)

model = LinearRegression()
model.fit(x_train,y_train)

print("\nSlope",model.coef_[0])
print("\nIntercept",model.intercept_)

y_pred = model.predict(x_test)
print("\nActual vs Predicted")
for actual, pred in zip(y_test,y_pred):
    print(f"Actual :{actual:.2f}, Predicted: {pred:.2f}")

mean_error = mean_absolute_error(y_pred,y_test)
r2 = r2_score(y_pred,y_test)

new_experience = pd.DataFrame([[9]],columns=['Experience_years'])
predicted_salary = model.predict(new_experience)
print(f"The salary for this {new_experience['Experience_years'][0]} is {predicted_salary}")

plt.scatter(x,y,color='red',label='Actual Data')
plt.plot(x,model.predict(x),color='blue',label='Regression Line')
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Salary Prediction')
plt.legend()
plt.show()
```
