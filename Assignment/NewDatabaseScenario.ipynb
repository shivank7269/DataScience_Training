```python
import numpy as np
import pandas as pd

employee = pd.read_csv("../Dataset/random/employee_attrition.csv")
sales = pd.read_csv("../Dataset/random/sample_data1.csv")
telecom = pd.read_csv("../Dataset/random/telecom_churn.csv")
```

```python
# EMPLOYEE DATA
print(f"{employee.head()}")
print(f"{employee.info()}")
print(f"{employee.describe()}")
# -> Spotting Null
print(f"Data That is Null :{employee.isnull().sum().sort_values(ascending=False)}")
# -> Fill Null
employee["Job_Satisfaction"]=employee["Job_Satisfaction"].fillna(round(employee["Job_Satisfaction"].mean()))
# -> Data Aggregation
print("Average Years at company according to job role\n")
print(employee.groupby('Job_Role')["Years_at_Company"].mean())
print("Average Income according to Department")
print(employee.groupby('Department')["Monthly_Income"].mean())

print("Attrition according to Department\n")
print(employee.groupby('Department')['Attrition'].value_counts())
print("Exporting Clean Data")

employee.to_csv("../Dataset/random/cleaned/employee.csv")
```

```python
# Sales Data
print(f"{sales.head()}")
print(f"{sales.info()}")
print(f"{sales.describe()}")
# check for null data
sales.isnull().sum()

# fill null value
sales["Monthly_Sales"] = sales["Monthly_Sales"].fillna(0)
# fill average performance rating if null
sales["Performance_Rating"] = sales["Performance_Rating"].fillna(int(sales["Performance_Rating"].mean()))

# Data aggregation
print("Average Monthly Salse according to department\n")
sales.groupby('Department')["Monthly_Sales"].mean()

print("Average Monthly salse according to Region\n")
sales.groupby('Region')['Monthly_Sales'].mean()

print("Average Salary according to region")
sales.groupby('Region')['Salary'].mean()
print("Exporting Clean Data")

sales.to_csv("../Dataset/random/cleaned/sales.csv")
```

```python
# Sales Data
print(f"{telecom.head()}")
print(f"{telecom.info()}")
print(f"{telecom.describe()}")
# check for null data
telecom.isnull().sum()

# fill null value
telecom["Churn"] = telecom["Churn"].fillna('Yes')

print(f"Number of churns {telecom["Churn"].value_counts()}")

print(f"{pd.crosstab(telecom['Tenure_Months'],telecom['Churn'])}")

print("Average Tenure before churning\n")
print(telecom.groupby('Churn')['Tenure_Months'].mean())

print("Exporting Clean Data")

telecom.to_csv("../Dataset/random/cleaned/telecom.csv")
```
