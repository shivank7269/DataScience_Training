```python
add_spend=np.array([2000,4000,6000,8000,10000,12000,14000,16000,18000])
sales = np.array([7000,15000,21000,25000,32000,36000,42000,47000,52000])

plt.scatter(add_spend,sales,color='blue',alpha=0.9)

z=np.polyfit(add_spend,sales,1)
p = np.poly1d(z)

plt.plot(add_spend,p(add_spend),'y--',label='Trend')  # 'r-- or y--' represent the line that cut thorough the points and its color

plt.xlabel("Ad Spend")
plt.ylabel("Sales")
plt.title('Ad Spend vs Sales')
plt.legend()           # show the label
plt.show()
```

# PROBABILITY

```python
import random

heads=0
total=10000

for i in range(total):
    result = random.randint(0,1)
    if result == 1:
        heads= heads+1
tails=total-heads

print(f"Heads :{heads}")
print(f"Tails :{tails}")
print(f"P(Head){heads/total}")
```

```python
heights = np.random.normal(165, 8, 200)
print("Shortest:", round(heights.min(), 1))
print("Tallest :", round(heights.max(), 1))
print("Average :", round(heights.mean(), 1))
plt.hist(heights, bins=50,
         color='#00BFA6',
         edgecolor='white')
plt.axvline(heights.mean(),
            color='red', linewidth=2,
            label='Average')

plt.title('Heights of 200 Students')
plt.xlabel('Height (cm)')
plt.ylabel('Number of Students')
plt.legend()

plt.show()
```

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np

n = 10
p = 0.25

p3 = binom.pmf(3, n, p)
print("P(exactly 3) =", round(p3, 3))

p5 = 1 - binom.cdf(4, n, p)
print("P(5 or more) =", round(p5, 3))

k = np.arange(0, n + 1)

plt.bar(k, binom.pmf(k, n, p),
        color='#7B2EBF',
        edgecolor='white')

plt.title('Probability of Correct Guesses (n=10, p=0.25)')
plt.xlabel('Number of Correct Answers')
plt.ylabel('Probability')
plt.xticks(k)

plt.show()
```
