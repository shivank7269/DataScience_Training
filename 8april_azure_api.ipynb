```python
import requests
import os
from dotenv import load_dotenv

load_dotenv()

# 🔗 Endpoint (Project / Foundry)
ENDPOINT = os.getenv('AZURE_ENDPOINT_GPT')

# 🔐 API Key from Azure AI Studio (Project Keys)
API_KEY = os.getenv("AZURE_API_KEY_GPT")

headers = {
    "Content-Type": "application/json",
    "api-key": API_KEY
}

payload = {
    "model": "gpt-4.1",
    "messages": [
        {"role": "system", "content": "You are a helpful AI assistant."},
        {"role": "user", "content": "What is the capital of France?"}
    ],
    "temperature": 0.7
}

response = requests.post(ENDPOINT, headers=headers, json=payload)

# 🧠 Extract only useful output
if response.status_code == 200:
    result = response.json()
    print(result["choices"][0]["message"]["content"])
else:
    print("Error:", response.status_code, response.text)
```

```python
# Scenario
# You’re training healthcare analysts.
# You want them to see how to connect securely to Azure OpenAI, ask questions about AI in healthcare, and then deploy the interaction into a Gradio interface so they can try it themselves.
```

```python
import gradio as gr
from dotenv import load_dotenv
import requests
import os

load_dotenv()

ENDPOINT = os.getenv("AZURE_ENDPOINT_GPT")
API_KEY = os.getenv("AZURE_API_KEY_GPT")

def ask_ai(prompt):
    headers = {
        "Content-Type": "application/json",
        "api-key": API_KEY
    }

    payload = {
        "model": "gpt-4.1",
        "messages": [
            {
                "role": "system",
                "content": "You are a healthcare AI assistant. Provide safe, non-diagnostic guidance."
            },
            {
                "role": "user",
                "content": prompt
            }
        ]
    }

    response = requests.post(ENDPOINT, headers=headers, json=payload)

    if response.status_code != 200:
        return f"Error: {response.text}"

    return response.json()["choices"][0]["message"]["content"]


# 🎛️ UI
demo = gr.Interface(
    fn=ask_ai,
    inputs=gr.Textbox(
        label="Ask about AI in Healthcare",
        placeholder="e.g., How is AI used in radiology?"
    ),
    outputs=gr.Textbox(label="AI Response"),
    title="🏥 AI in Healthcare Assistant",
    description="Explore how AI supports healthcare decisions (educational use only)."
)

demo.launch()
```
