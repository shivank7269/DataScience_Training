Scenario: Automated Business Report Assistant
🎯 Problem Statement
In companies:
- Managers spend hours compiling data from multiple sources (CRM, ERP, spreadsheets).
- Reports often require manual formatting, chart creation, and consistency checks.
- This slows down decision-making and introduces human error.
👉 Solution
A multi-agent AutoGen system where:
- 🤖 Assistant Agent drafts report templates and queries relevant data.
- ⚙️ UserProxy Agent executes data pulls and formatting automatically.
- 🔁 The system iterates until the report meets formatting, accuracy, and completeness standards.
🔄 Workflow
Business Request → AI drafts report structure → Data fetched & formatted → Charts auto-generated → Validation loop → Final polished report

```python
import os
import autogen
from dotenv import load_dotenv

load_dotenv()
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

config_list = [
    {
        "model": "llama-3.3-70b-versatile",
        "api_key": os.environ["GROQ_API_KEY"],
        "base_url": "https://api.groq.com/openai/v1"
    }
]

llm_config = {"config_list": config_list, "temperature": 0}

planner = autogen.AssistantAgent(
    name="Planner",
    llm_config=llm_config,
    system_message="Create a structured business report outline with sections like summary, metrics, insights."
)

analyst = autogen.AssistantAgent(
    name="Analyst",
    llm_config=llm_config,
    system_message="Write Python code to analyze data and generate insights with statistics and charts."
)

reviewer = autogen.AssistantAgent(
    name="Reviewer",
    llm_config=llm_config,
    system_message="Review the report for accuracy, clarity, and completeness. Suggest improvements if needed."
)

executor = autogen.UserProxyAgent(
    name="Executor",
    human_input_mode="NEVER",
    max_consecutive_auto_reply=8,
    code_execution_config={"work_dir": "coding", "use_docker": False}
)

task = """
Sales Data: [12000, 15000, 10000, 18000, 13000]

Steps:
1. Create a business report structure
2. Analyze the data (mean, median, mode, min, max)
3. Generate insights
4. Create a simple chart
5. Review and refine the report
"""

groupchat = autogen.GroupChat(
    agents=[planner, analyst, reviewer, executor],
    messages=[],
    max_round=10
)

manager = autogen.GroupChatManager(
    groupchat=groupchat,
    llm_config=llm_config
)

executor.initiate_chat(manager, message=task)
```

Scenario: AI Code Review & Execution System (Corporate Dev Team)
🎯 Problem Statement

In companies:

Developers write code

Reviewers check for bugs

Testing is manual

👉 Solution:
A multi-agent system where:

👨‍💻 Coder Agent writes code

🔍 Reviewer Agent validates + executes code

🔁 Loop continues until correct

```python
import os
import autogen
from dotenv import load_dotenv
load_dotenv()

# -----------------------------
# Groq API Key
# -----------------------------
groq_api_key = os.getenv("GROQ_API_KEY")
os.environ["GROQ_API_KEY"] = groq_api_key

# -----------------------------
# LLM Config (Groq)
# -----------------------------
llm_config = {
    "config_list": [
        {
            "model": "llama-3.3-70b-versatile",
            "api_key": groq_api_key,
            "base_url": "https://api.groq.com/openai/v1"
        }
    ],
    "temperature": 0
}

# -----------------------------
# Agent 1: Coder
# -----------------------------
coder = autogen.AssistantAgent(
    name="Coder",
    llm_config=llm_config,
    system_message="""You are an expert Python developer.

Write clean, well-commented code.

Rules:
- Handle edge cases
- Add test cases
- Ensure code runs without errors
"""
)

# ----------------------------
# Agent 2: Reviewer (Executor)
# -----------------------------
reviewer = autogen.UserProxyAgent(
    name="Reviewer",
    human_input_mode="NEVER",
    code_execution_config={
        "work_dir": "review",
        "use_docker": False,
        "last_n_messages": 2
    },
    is_termination_msg=lambda x: "APPROVED" in (x.get("content") or ""),
    system_message="""You are a strict code reviewer.

Rules:
1. Execute ONLY python code
2. If execution succeeds → say ONLY: APPROVED
3. If error → ask for fix
4. Do NOT explain anything"""
)

# -----------------------------
# Start Collaboration
# -----------------------------
reviewer.initiate_chat(
    coder,
    message="""Create a Python function that:
- Generates Fibonacci sequence up to N
- Returns it as a list
- Handles edge cases (N=0, negative N)

Test it with N=10 and print output"""
)
```

```python
import os
import autogen
from dotenv import load_dotenv

load_dotenv()

os.environ["AUTOGEN_USE_DOCKER"] = "False"

groq_api_key = os.getenv("GROQ_API_KEY")
if not groq_api_key:
    raise ValueError("GROQ_API_KEY not found in .env")

llm_config = {
    "config_list": [
        {
            "model": "llama-3.1-8b-instant",
            "api_key": groq_api_key,
            "base_url": "https://api.groq.com/openai/v1"
        }
    ],
    "temperature": 0.2
}

pm = autogen.AssistantAgent(
    name="ProductManager",
    llm_config=llm_config,
    system_message="Break the feature request into clear development tasks."
)

coder = autogen.AssistantAgent(
    name="Coder",
    llm_config=llm_config,
    system_message="Write clean Python FastAPI code with CRUD operations. Only output code."
)

reviewer = autogen.AssistantAgent(
    name="Reviewer",
    llm_config=llm_config,
    system_message="Review code for bugs, PEP8 compliance, and improvements. Suggest fixes if needed."
)

tester = autogen.AssistantAgent(
    name="Tester",
    llm_config=llm_config,
    system_message="Write pytest unit tests and validate the code."
)

user = autogen.UserProxyAgent(
    name="User",
    human_input_mode="NEVER",
    code_execution_config={
        "work_dir": "project",
        "use_docker": False
    }
)

groupchat = autogen.GroupChat(
    agents=[user, pm, coder, reviewer, tester],
    messages=[],
    max_round=15,
    speaker_selection_method="round_robin"
)

manager = autogen.GroupChatManager(
    groupchat=groupchat,
    llm_config=llm_config
)

user.initiate_chat(
    manager,
    message="""
Build a FastAPI TODO REST API with:
- CRUD operations
- In-memory storage
- Proper error handling
- Pytest unit tests
- Follow PEP8 standards
"""
)
```

Scenario: AI-Powered HR Onboarding System (Corporate HR Team)
🎯 Problem Statement
In companies:
- HR teams manually prepare onboarding documents and checklists.
- Managers validate compliance with policies.
- Employees often face delays in receiving resources and training schedules.
👉 Solution
A multi-agent system where:
- 📝 HR Agent prepares onboarding documents, training schedules, and resource lists.
- ✅ Compliance Agent reviews documents for policy adherence and legal requirements.
- 📂 Resource Agent allocates accounts, devices, and access permissions.
- 🔁 The loop continues until all onboarding steps are validated and complete.
🔄 Workflow
Onboarding Request → HR Agent drafts documents → Compliance Agent validates → Resource Agent provisions accounts/devices → Iteration loop → Final approved onboarding package
🧹 Conflict Resolution
- Resolve conflicting policy requirements (e.g., different regional compliance rules).
- Standardize onboarding templates across departments.
- Ensure system avoids duplicate resource allocation (e.g., multiple accounts for same employee).

```python
import os
import autogen
from dotenv import load_dotenv

load_dotenv()

os.environ["AUTOGEN_USE_DOCKER"] = "False"

groq_api_key = os.getenv("GROQ_API_KEY")
if not groq_api_key:
    raise ValueError("GROQ_API_KEY not found")

llm_config = {
    "config_list": [
        {
            "model": "llama-3.1-8b-instant",
            "api_key": groq_api_key,
            "base_url": "https://api.groq.com/openai/v1"
        }
    ],
    "temperature": 0.3,
    "max_tokens": 600
}

# 📝 HR Agent
hr_agent = autogen.AssistantAgent(
    name="HRAgent",
    llm_config=llm_config,
    system_message=(
        "Create structured onboarding documents including:\n"
        "- Employee details\n- Training schedule\n- Resource checklist\n"
        "Keep it concise and complete."
    )
)

# ✅ Compliance Agent
compliance_agent = autogen.AssistantAgent(
    name="ComplianceAgent",
    llm_config=llm_config,
    system_message=(
        "Review onboarding content and respond in this format:\n"
        "1. Policy Issues\n2. Missing Items\n3. Fix Suggestions\n"
        "If everything is correct, reply only with 'APPROVED'."
    )
)

# 📂 Resource Agent
resource_agent = autogen.AssistantAgent(
    name="ResourceAgent",
    llm_config=llm_config,
    system_message=(
        "Allocate resources (email, GitHub, laptop, access).\n"
        "Ensure no duplicates and completeness."
    )
)

# 👤 Executor (User Proxy)
user = autogen.UserProxyAgent(
    name="User",
    human_input_mode="NEVER",
    code_execution_config={
        "work_dir": "hr_project",
        "use_docker": False
    }
)

# 🔄 Termination Condition
def is_termination(msg):
    if msg is None:
        return False
    content = msg.get("content", "")
    return "APPROVED" in content

# 🔄 Group Chat (MAX 3 ROUNDS)
groupchat = autogen.GroupChat(
    agents=[user, hr_agent, compliance_agent, resource_agent],
    messages=[],
    max_round=3,
    speaker_selection_method="round_robin"
)

# 🎯 Manager
manager = autogen.GroupChatManager(
    groupchat=groupchat,
    llm_config=llm_config,
    is_termination_msg=is_termination
)

# 🚀 Start Workflow
user.initiate_chat(
    manager,
    message="""
Onboard a new Software Engineer in India:
- Generate onboarding documents
- Include training schedule
- Ensure compliance with company policies
- Allocate required resources (email, GitHub, laptop)
- Avoid duplicate allocations
"""
)
```

You are part of an AI engineering team building an AutoGen-based developer assistant.

The system should:

Accept a user problem statement
Generate code
Review the code
Execute it
Provide feedback and corrections

This must be implemented using multiple collaborating agents.

🔹 Problem Statement

A user gives the following input:

“Write a Python function to find the second largest number in a list and handle edge cases.”

Your AutoGen system should:

Generate correct code
Validate logic
Execute the code
Handle errors
Provide final response
🔹 Requirements
✅ 1. Agent Design (MANDATORY)

Create at least 3 agents:

🔸 Coder Agent
Generates Python code
Should follow best practices
Must handle edge cases
🔸 Reviewer Agent
Reviews code for:
Bugs
Logic errors
Edge cases
Suggests improvements
🔸 Executor Agent (UserProxy)
Runs the code
Returns output/errors
🔹 2. Agent Collaboration Flow

Design the flow such that:

User → Coder → Reviewer → Executor → Reviewer (if error) → Final Output

👉 The system should not stop after one step
👉 It must iterate until correct output is achieved

🔹 3. Functional Expectations

Your system must:

✔ Handle incorrect code generated initially
✔ Ensure reviewer gives structured feedback
✔ Prevent infinite loops
✔ Show final clean output

🔹 4. Technical Constraints
Use AutoGen framework
Use LLM-based agents
Enable code execution
Add termination condition
🔹 5. Edge Cases to Handle

Test your system with:

Empty list
List with one element
Duplicate values
Negative numbers

```python
import os
import autogen
from dotenv import load_dotenv

load_dotenv()

# Disable Docker completely
os.environ["AUTOGEN_USE_DOCKER"] = "False"

groq_api_key = os.getenv("GROQ_API_KEY")
if not groq_api_key:
    raise ValueError("Missing GROQ_API_KEY")

llm_config = {
    "config_list": [
        {
            "model": "llama-3.1-8b-instant",
            "api_key": groq_api_key,
            "base_url": "https://api.groq.com/openai/v1"
        }
    ],
    "temperature": 0.2,
    "max_tokens": 600
}

coder = autogen.AssistantAgent(
    name="Coder",
    llm_config=llm_config,
    system_message=(
        "Write Python code only. Solve the problem correctly. "
        "Handle edge cases: empty list, single element, duplicates, negative numbers. "
        "Include test cases and print outputs."
    )
)

reviewer = autogen.AssistantAgent(
    name="Reviewer",
    llm_config=llm_config,
    system_message=(
        "Review the code and give structured feedback:\n"
        "1. Bugs\n2. Edge Cases\n3. Improvements\n"
        "If everything is correct, reply only with 'APPROVED'."
    )
)


executor = autogen.UserProxyAgent(
    name="Executor",
    human_input_mode="NEVER",
    code_execution_config={
        "work_dir": "code_project",
        "use_docker": False
    }
)

def is_termination(msg):
    if msg is None:
        return False
    content = msg.get("content", "")
    return "APPROVED" in content

groupchat = autogen.GroupChat(
    agents=[executor, coder, reviewer],
    messages=[],
    max_round=3,
    speaker_selection_method="round_robin"
)

manager = autogen.GroupChatManager(
    groupchat=groupchat,
    llm_config=llm_config,
    is_termination_msg=is_termination
)

executor.initiate_chat(
    manager,
    message="""
Write a Python function to find the second largest number in a list.

Requirements:
- Handle empty list
- Handle single element list
- Handle duplicates
- Handle negative numbers
- Include test cases
"""
)
```
